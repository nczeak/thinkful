# Distribution 1: Gamma Distribution

```python
gammaDistributionArray = np.random.gamma(0.5, 0.5, 100)
plotFunctions.hist(gammaDistributionArray)
plotFunctions.axvline(gammaDistributionArray.mean(), color='r', linestyle='solid', linewidth=2)
plotFunctions.axvline((gammaDistributionArray.mean() + gammaDistributionArray.std()), color='g', linestyle='dashed', linewidth=2)
plotFunctions.axvline((gammaDistributionArray.mean() - gammaDistributionArray.std()), color='g', linestyle='dashed', linewidth=2)

'''Create labels for the plot, display the plot, and print descriptive statistics'''
plotFunctions.title("100 Gamma Distributed Numbers")
plotFunctions.xlabel("Generated Numbers")
plotFunctions.ylabel("Frequency of Generated Numbers")
plotFunctions.show()
print("Mean of the distribution: " + str(np.mean(gammaDistributionArray)))
print("Median of the distribution: " + str(np.median(gammaDistributionArray)))
print("Mode of the distribution: " + str(stats.mode(gammaDistributionArray)))
print("Standard Deviation of the distribution: " + str(np.std(gammaDistributionArray)))
print("Standard Error of the distribution: " + str(stats.sem(gammaDistributionArray)))
print("Variance of the distribution: " + str(np.var(gammaDistributionArray)))
```

In this example, I'd use the median as the measurement of central tendency. The mean is heavily skewed by the high amount of values < 0.1. Because of that, I'd rather use median to describe where 50% of the data is split.

# Additionally:

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.

Add them together to create a third variable.

Graph the third variable using a histogram.

Compute the mean and standard deviation and plot them as vertical lines on the histogram.

Evaluate the descriptive statistics against the data.

```python
firstDistribution = np.random.normal(5, 0.5, 10000)
secondDistribution = np.random.normal(10, 1, 10000)
totalDistribution = firstDistribution + secondDistribution

plotFunctions.hist(totalDistribution)
plotFunctions.axvline(totalDistribution.mean(), color='r', linestyle='solid', linewidth=2)
plotFunctions.axvline((totalDistribution.mean() + totalDistribution.std()), color='g', linestyle='dashed', linewidth=2)
plotFunctions.axvline((totalDistribution.mean() - totalDistribution.std()), color='g', linestyle='dashed', linewidth=2)
plotFunctions.show()
print("Mean of the distribution: " + str(np.mean(totalDistribution)))
print("Median of the distribution: " + str(np.median(totalDistribution)))
print("Mode of the distribution: " + str(stats.mode(totalDistribution)))
print("Standard Deviation of the distribution: " + str(np.std(totalDistribution)))
print("Standard Error of the distribution: " + str(stats.sem(totalDistribution)))
print("Variance of the distribution: " + str(np.var(totalDistribution)))
```

The two normal distributions added together form a single normal distribution. The standard deviation is 1, which seems to reflect the larger standard deviation of the two distributions. The mean is 15, which is simply the mean of 5 added to the mean of 10.
