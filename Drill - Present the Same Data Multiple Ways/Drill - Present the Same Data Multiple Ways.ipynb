```python
'''Import necessary modules'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plotFunctions
from scipy import stats
import seaborn as sns
%matplotlib inline
```

# Drill 1:

```python
dataFrame = pd.read_csv("QBStats_all.csv")
```

```python
plotFunctions.hist(dataFrame["yds"], density = True)
plotFunctions.show()
```

```python
plotFunctions.plot(sorted(np.random.normal(dataFrame["yds"].mean(), dataFrame["yds"].std(), len(dataFrame["yds"]))), sorted(dataFrame["yds"]), "o")
plotFunctions.show()
```

```python
plotFunctions.scatter(dataFrame["att"], dataFrame["yds"])
plotFunctions.show()
```

```python
averageList = []
for thisYear in range(1996, 2017):
    averageList.append(dataFrame[dataFrame["year"] == thisYear]["yds"].mean())
plotFunctions.plot(sorted(set(dataFrame["year"])), averageList)
plotFunctions.show()
```
