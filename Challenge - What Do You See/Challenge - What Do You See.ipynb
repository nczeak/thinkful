# Evolution of the Passing Game in the NFL

#### Introduction:

Ever since the invention of the forward pass in 1906, football has been slowly trending away from the run and towards the choatic and explosive pass. But just how much more has the pass been used in recent years, and how much better are the modern quarterbacks than the gunslingers of yesteryear? This project explores just how much the passing game has evolved in the NFL.

First, I'll import all of the modules I will use.

```python
'''Import necessary modules'''
import matplotlib.pyplot as plotFunctions
import numpy as np
import pandas as pd
%matplotlib inline
```

Next, I'll grab the data frame used in this project from the CSV I downloaded from Kaggle.

```python
'''Create a data frame from the downloaded CSV'''
quarterbackDataFrame = pd.read_csv("QBStats_all.csv")
```

#### Evolution of the Use of a Quarterback:

For my first visualization, I want to see just how much more often quarterbacks are being used in the NFL in today's game. I decided to plot two histograms from 1996 and 2016 that show the average pass attempts per game. This is displayed in Figure 1.

```python
'''Create histograms for both the passing attempts in 1996 and 2016'''
plotFunctions.hist(quarterbackDataFrame[quarterbackDataFrame["year"] == 1996]["att"], color = "blue", density = True, alpha = 0.5)
plotFunctions.hist(quarterbackDataFrame[quarterbackDataFrame["year"] == 2016]["att"], color = "red", density = True, alpha = 0.5)

'''Label the plot and display it'''
plotFunctions.title("Figure 1: Pass Attempts per Game, 1996 & 2016")
plotFunctions.xlabel("Pass Attempts")
plotFunctions.ylabel("Normalzied Probability of Pass Attempts")
plotFunctions.legend(("1996, mean = " + str(round(quarterbackDataFrame[quarterbackDataFrame["year"] == 1996]["att"].mean(), 1)), "2016, mean = " + str(round(quarterbackDataFrame[quarterbackDataFrame["year"] == 2016]["att"].mean(), 1))))
plotFunctions.show()
```

As shown in Figure 1, 2016 had a higher mean of pass attempts per game. The distribution also had more pass attempts in the 50-60 range than 1996 did. This clearly shows that quarterbacks are being used more in the modern NFL than in the previous era.

#### Evolution of Total Contribution of a Quarterback per Game:

I wanted my second vizualization to show the change of the total contribution of quarterbacks per game. I decided to use passing yards per game, as displayed in Figure 2.

```python
'''Create the first plot for 1996 and label it'''
plotFunctions.figure(figsize = (10, 5))
plotFunctions.subplot(1, 2, 1)
plotFunctions.boxplot(quarterbackDataFrame[quarterbackDataFrame["year"] == 1996]["yds"])
plotFunctions.title("Figure 2a: Pass Yards per Game 1996")
plotFunctions.ylabel("Pass Yards")

'''Create the second plot for 2016 and label it'''
plotFunctions.subplot(1, 2, 2)
plotFunctions.boxplot(quarterbackDataFrame[quarterbackDataFrame["year"] == 2016]["yds"])
plotFunctions.title("Figure 2b: Pass Yards per Game 2016")
plotFunctions.ylabel("Pass Yards")

'''Show the plots and print any relevant information'''
plotFunctions.show()
print("Median Pass Yards 1996: " + str(quarterbackDataFrame[quarterbackDataFrame["year"] == 1996]["yds"].median()))
print("Median Pass Yards 2016: " + str(quarterbackDataFrame[quarterbackDataFrame["year"] == 2016]["yds"].median()))
```

Figure 2a (representing 1996) shows a box with a median of 183 pass yards in a game, and Figure 2b (representing 2016) shows a box with a median of 234 pass yards in a game. The 75% range is also slightly higher in 2016 than in 1996. This means that quarterbacks are not only being used more; quarterbacks are also contributing more to each game.

#### Evolution of Quarterback Efficiency:

While total numbers are a good indication of total use of a quarterback, I also wanted to see how much better a quarterback is per throw. This efficiency can be described using Quarterback Rating. Quarterback Rating (or QBR) is a measurement of how efficient a quarterback is with each throw. Figure 3 shows the average QBR in a game per year.

```python
'''Create a list for average QBR per year. Loop through the years and populate the list'''
averageList = []
for thisYear in range(1996, 2017):
    averageList.append(quarterbackDataFrame[quarterbackDataFrame["year"] == thisYear]["rate"].mean())
    
'''Create the plot, label it, and display it'''
plotFunctions.plot(sorted(set(quarterbackDataFrame["year"])), averageList)
plotFunctions.title("Figure 3: Average QB Rating per Game, 1996-2016")
plotFunctions.xlabel("Year")
plotFunctions.ylabel("Average QB Rating per Game")
plotFunctions.show()
```

As Figure 3 shows, the average QBR per game has increased since 1996. While average QBR started around 76, it has trended towards the high 80s in the span of a couple decades. This shows that quarterbacks are becoming much more efficient with their passes.

#### Visualizing the Total Evolution of Passing in NFL Games:

For my last visualization, I wanted to show what games look like with both total yards and efficiency numbers. Figure 4 shows the QBR vs. Yards in each game from both the 1996 and 2016 seasons.

```python
'''Create the first plot for 1996 and label it'''
plotFunctions.figure(figsize = (10, 5))
plotFunctions.subplot(1, 2, 1)
plotFunctions.scatter(quarterbackDataFrame[quarterbackDataFrame["year"] == 1996]["rate"], quarterbackDataFrame[quarterbackDataFrame["year"] == 1996]["yds"], color = "red")
plotFunctions.title("Figure 4a: QBR vs. Yards 1996")
plotFunctions.xlabel("Quarterback Rating")
plotFunctions.ylabel("Yards")

'''Create the second plot for 2016 and label it'''
plotFunctions.subplot(1, 2, 2)
plotFunctions.scatter(quarterbackDataFrame[quarterbackDataFrame["year"] >= 2016]["rate"], quarterbackDataFrame[quarterbackDataFrame["year"] == 2016]["yds"], color = "blue")
plotFunctions.title("Figure 4b: QBR vs. Yards 2016")
plotFunctions.xlabel("Quarterback Rating")
plotFunctions.ylabel("Yards")

'''Show both plots'''
plotFunctions.show()
```

The main cluster of games in 1996 shows a lower average QBR and total passing yards than the main cluster of games from 2016. This means that in 2016, you're more likely to see a game with a higher QBR and higher total yards than you'd see in a game from 1996.

#### Conclusion:

In conclusion, passing in the NFL has changed drastically over the past two decades. Quarterbacks are being used more, posting higher total numbers, and becoming much more efficient with each pass made.
